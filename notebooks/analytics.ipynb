## üìä 1. Ph√¢n T√≠ch Xu H∆∞·ªõng Seasonal v√† Events

````python
# Ph√¢n t√≠ch seasonal trends v√† events
def analyze_seasonal_trends():
    # T·∫°o mapping cho m√πa v√† events
    current_month = datetime.now().month
    current_season = ""
    upcoming_events = []
    
    # X√°c ƒë·ªãnh m√πa hi·ªán t·∫°i
    if current_month in [12, 1, 2]:
        current_season = "ƒê√¥ng"
        upcoming_events = ["Noel", "T·∫øt Nguy√™n ƒê√°n", "Valentine"]
    elif current_month in [3, 4, 5]:
        current_season = "Xu√¢n"  
        upcoming_events = ["8/3", "Gi·ªó t·ªï H√πng V∆∞∆°ng", "30/4-1/5"]
    elif current_month in [6, 7, 8]:
        current_season = "H√®"
        upcoming_events = ["Ngh·ªâ h√®", "Trung thu s·ªõm"]
    else:  # [9, 10, 11]
        current_season = "Thu"
        upcoming_events = ["Trung thu", "Halloween", "Black Friday"]
    
    print(f"üóìÔ∏è Hi·ªán t·∫°i: Th√°ng {current_month} - M√πa {current_season}")
    print(f"üéâ S·ª± ki·ªán s·∫Øp t·ªõi: {', '.join(upcoming_events)}")
    
    return current_season, upcoming_events

current_season, upcoming_events = analyze_seasonal_trends()
````

````python
# Ph√¢n t√≠ch data theo seasonal patterns
def extract_seasonal_insights(dataframes):
    insights = {}
    
    # T·ª´ YouTube trends data
    if 'youtube_bakery_gaming_trends_cleaned.csv' in dataframes:
        youtube_df = dataframes['youtube_bakery_gaming_trends_cleaned.csv']
        
        # Ph√¢n t√≠ch keywords theo m√πa
        halloween_keywords = youtube_df[youtube_df['tu_khoa_tim_kiem'].str.contains('halloween|pumpkin|orange|spooky', case=False, na=False)]
        insights['halloween_trends'] = {
            'total_videos': len(halloween_keywords),
            'avg_views': halloween_keywords['luot_xem'].mean() if not halloween_keywords.empty else 0,
            'popular_keywords': halloween_keywords['tu_khoa_tim_kiem'].value_counts().head(5).to_dict()
        }
        
        # Ph√¢n t√≠ch theo nh√≥m ƒë·ªëi t∆∞·ª£ng
        target_groups = youtube_df['nhom_doi_tuong'].value_counts()
        insights['target_groups'] = target_groups.to_dict()
    
    # T·ª´ consumer trends data  
    consumer_files = [f for f in dataframes.keys() if 'consumer' in f.lower()]
    if consumer_files:
        consumer_df = dataframes[consumer_files[0]]
        
        # Ph√¢n t√≠ch theo seasonal preferences
        if 'season' in consumer_df.columns:
            seasonal_prefs = consumer_df.groupby('season')['trending_flavors'].apply(list)
            insights['seasonal_flavors'] = seasonal_prefs.to_dict()
    
    return insights

seasonal_insights = extract_seasonal_insights(dataframes)
print("üîç Seasonal Insights:", seasonal_insights)
````

## ü§ñ 3. Feature Engineering cho ML/DL Models

````python
# Feature Engineering cho training data
def create_training_features(dataframes):
    """
    T·∫°o features ƒë·ªÉ train model d·ª± ƒëo√°n xu h∆∞·ªõng v√† g·ª£i √Ω c√¥ng th·ª©c
    """
    training_data = []
    
    # T·ª´ YouTube data
    if 'youtube_bakery_gaming_trends_cleaned.csv' in dataframes:
        youtube_df = dataframes['youtube_bakery_gaming_trends_cleaned.csv']
        
        for _, row in youtube_df.iterrows():
            feature_row = {
                # Time features
                'month': datetime.now().month,
                'season': current_season,
                'is_weekend': datetime.now().weekday() >= 5,
                
                # Content features
                'video_category': row.get('chu_de', ''),
                'search_keyword': row.get('tu_khoa_tim_kiem', ''),
                'target_group': row.get('nhom_doi_tuong', ''),
                'age_group': row.get('do_tuoi', ''),
                
                # Engagement features
                'views': row.get('luot_xem', 0),
                'likes': row.get('luot_thich', 0),
                'comments': row.get('luot_binh_luan', 0),
                'engagement_rate': row.get('engagement_rate_%', 0),
                'viral_level': row.get('muc_do_viral', ''),
                
                # Food features
                'detected_foods': row.get('thuc_pham_phat_hien', ''),
                'preferred_foods': row.get('thuc_pham_yeu_thich', ''),
                'detected_desserts': row.get('banh_ngot_phat_hien', ''),
                'preferred_desserts': row.get('banh_ngot_yeu_thich', ''),
                
                # Target variables (ƒë·ªÉ predict)
                'popularity_score': row.get('diem_noi_tieng', 0),
                'is_trending': row.get('muc_do_viral', '') == 'C·ª±c k·ª≥ viral (100.0)'
            }
            training_data.append(feature_row)
    
    training_df = pd.DataFrame(training_data)
    
    # Th√™m seasonal features
    training_df['upcoming_events'] = str(upcoming_events)
    training_df['halloween_season'] = training_df['month'] == 10
    
    return training_df

# T·∫°o training dataset
training_features = create_training_features(dataframes)
print("üìä Training Dataset Shape:", training_features.shape)
print("\nüîß Features Created:")
print(training_features.columns.tolist())
print("\nüìã Sample Data:")
display(training_features.head())
````
