```python
import pandas as pd
import glob
import os

# Đường dẫn tới folder data/raw (cần chỉnh lại cho đúng với đường dẫn thực tế)
folder_path = "../data/raw/"
all_csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Đọc tất cả file CSV, bỏ vào list, và nối (concat) chúng lại thành 1 DataFrame lớn
df_list = [pd.read_csv(f) for f in all_csv_files]
df_all = pd.concat(df_list, ignore_index=True)

print(df_all.head())  # Xem thử 5 dòng đầu, hoặc xử lý tiếp
```
